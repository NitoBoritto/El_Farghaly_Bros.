```python
sns.countplot(x = 'y',data = df)
plt.title('Target Variable Distribution (y)')
plt.show()
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

plt.figure(figsize=(16, 18))
for i, col in enumerate(num_cols, 1):
    plt.subplot(4, 3, i)
    sns.boxplot(x='y', y=col, data=df, palette='Set2')
    plt.title(f'Outliers in {col}', fontsize=12, fontweight='bold')
    plt.xlabel('Target (y)')
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

```python
categorical_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan']

for col in categorical_cols:
    print(f"\n📊 Cross-tabulation: {col.upper()} vs Y")
    table = pd.crosstab(df[col], df['y'], margins=True, margins_name="Total")
    
    display(table)
    print("-" * 50) 
```

```python
categorical_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan']

for col in categorical_cols:
    print(f"\n📊 Cross-tabulation: {col.upper()} vs Y")
    table = pd.crosstab(df[col], df['y'], normalize='index') * 100
    
    display(table)
    print("-" * 50) 
```
