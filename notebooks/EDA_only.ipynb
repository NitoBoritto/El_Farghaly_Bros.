```python
sns.countplot(x = 'y',data = df)
plt.title('Target Variable Distribution (y)')
plt.show()
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

plt.figure(figsize=(16, 18))
for i, col in enumerate(num_cols, 1):
    plt.subplot(4, 3, i)
    sns.boxplot(x='y', y=col, data=df, palette='Set2')
    plt.title(f'Outliers in {col}', fontsize=12, fontweight='bold')
    plt.xlabel('Target (y)')
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

```python
categorical_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan']

for col in categorical_cols:
    print(f"\n📊 Cross-tabulation: {col.upper()} vs Y")
    table = pd.crosstab(df[col], df['y'], margins=True, margins_name="Total")
    
    display(table)
    print("-" * 50) 
```

```python
categorical_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan']

for col in categorical_cols:
    print(f"\n📊 Cross-tabulation: {col.upper()} vs Y")
    table = pd.crosstab(df[col], df['y'], normalize='index') * 100
    
    display(table)
    print("-" * 50) 
```

```python
num_cols = df.select_dtypes(include=['int64','float64'])
corr_matrix = num_cols.corr()

plt.figure(figsize=(12,6))
sns.heatmap(
    corr_matrix,
    annot = True,
    cmap = 'coolwarm',
    fmt = '.2f',
    linewidths = 0.5
)
plt.title('Correlation Matrix of Numerical Features', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# تحديد الأعمدة التي نريد رسمها
categorical_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan']

# تجهيز لوحة كبيرة مقسمة لـ 3 صفوف وعمودين
fig, axes = plt.subplots(3, 2, figsize=(15, 18))
axes = axes.flatten()

# حلقة تكرارية لرسم كل عمود
for i, col in enumerate(categorical_cols):
    sns.countplot(x=col, hue='y', data=df, ax=axes[i], palette='Set1')
    axes[i].set_title(f'{col.capitalize()} vs Target (y)', fontsize=14, fontweight='bold')
    axes[i].tick_params(axis='x', rotation=45) # لتدوير أسماء الفئات حتى لا تتداخل
    axes[i].set_ylabel('Count')

# ترتيب المسافات وعرض الرسم
plt.tight_layout()
plt.show()
```

```python
for col in num_cols:
    plt.figure(figsize=(8, 5)) # عشان يفتح رسمة جديدة لكل عمود
    
    # هنا حطينا col بدل num_cols
    sns.kdeplot(data=df, x=col, hue='y', fill=True, palette='Set2')
    
    plt.title(f'Distribution of {col} by Target') # هيخلي العنوان باسم العمود
    plt.show()
    print("-" * 50)
```

```python
for col in categorical_cols:
    sns.countplot(x= col, hue='y', data=df, palette='Set1')
    plt.xticks(rotation=45) # Rotate labels if they are too long
    plt.title('Categorical Feature vs Target')
    plt.show()
    print("-" * 50)
```

```python
cols_to_plot = ['age', 'duration', 'euribor3m', 'y']

# رسم الـ Pairplot
sns.pairplot(df[cols_to_plot], hue='y', palette='Set1')
plt.show()
```

```python
# 1. Convert 'unknown' strings to actual NaN values so the library can detect them
df.replace('unknown', np.nan, inplace=True)

# ==========================================
# Chart 1: Bar Chart
# Shows the total count of non-missing values per column
# ==========================================
msno.bar(df, figsize=(12, 6), color="dodgerblue", fontsize=12)
plt.title('Count of Complete Data per Column', fontsize=16)
plt.show()

# ==========================================
# Chart 2: Matrix Plot
# Excellent for seeing WHERE the missing values are and if they are related
# ==========================================
msno.matrix(df, figsize=(12, 6), color=(0.2, 0.4, 0.6), sparkline=False)
plt.title('Missing Data Map (White lines indicate missing values)', fontsize=16)
plt.show()
```
